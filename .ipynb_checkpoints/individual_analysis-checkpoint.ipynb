```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_sessions = pd.read_csv(filepath_or_buffer = './data/ga_sessions.csv')
df_sessions.head()
```

```python
df = pd.read_csv(filepath_or_buffer = "./data/ga_hits.csv")
total_dataframe = pd.merge(left = df, right = df_sessions, on = "session_id")
total_dataframe.head(10)
```

```python
correlation_referer = total_dataframe[['hit_referer', 'event_action']]
correlation_referer.head(10)
```

```python
pure_df = correlation_referer.dropna()
pure_df.head(10)
```

```python
pure_df['hit_referer'].value_counts()
```

```python
len(pure_df['hit_referer'].unique())
```

```python
from sklearn import preprocessing 
label = preprocessing.LabelEncoder()
data_encoded = pd.DataFrame()

for i in pure_df.columns:
    data_encoded[i] = label.fit_transform(pure_df[i])
```

```python
test_data = data_encoded.head(140000)
```

```python
test_data_2 = data_encoded.tail(140000)
```

```python
from scipy.stats import chi2_contingency 
import numpy as np 

def cramers_V(var1, var2):
    crosstab = np.array(pd.crosstab(var1, var2, rownames = None, colnames = None))
    stat = chi2_contingency(crosstab)[0]
    obs = np.sum(crosstab)#Number of observations 
    mini = min(crosstab.shape) - 1
    return np.sqrt(stat/(obs*mini))
```

```python
rows= []

for var1 in test_data:
  col = []
  for var2 in test_data :
    cramers =cramers_V(test_data[var1], test_data[var2]) # Cramer's V test
    col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
  rows.append(col)
  
cramers_results = np.array(rows)
df = pd.DataFrame(cramers_results, columns = data_encoded.columns, index =data_encoded.columns)

df
```

```python
rows= []

for var1 in test_data_2:
  col = []
  for var2 in test_data_2:
    cramers =cramers_V(test_data_2[var1], test_data_2[var2]) # Cramer's V test
    col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
  rows.append(col)
  
cramers_results = np.array(rows)
df = pd.DataFrame(cramers_results, columns = data_encoded.columns, index =data_encoded.columns)

df
```
